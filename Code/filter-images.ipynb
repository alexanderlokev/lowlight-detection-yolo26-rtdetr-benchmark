```python
import os
import shutil
import random
```

```python
base_dir = "CCPD2019/main/file/location"
filter_dir = "CCPD2019/filtered/file/location"
os.makedirs(filter_dir, exist_ok=True)
images = os.listdir(base_dir)
```

```python
# Define Image Characteristics
def parse_ccpd_name(name):
    parts = name.replace(".jpg", "").split("-")

    area = parts[0]
    tilt_degree = parts[1]
    bounding_box_coordinate = parts[2]
    four_vertices_location = parts[3]
    licence_plate_number = parts[4]
    brightness = parts[5]
    blurriness = parts[6]

    return {
        "area" : area,
        "tilt_degree" : tilt_degree,
        "bounding_box_coordinate" : bounding_box_coordinate,
        "four_vertices_location" : four_vertices_location,
        "licence_plate_number" : licence_plate_number,
        "brightness" : brightness,
        "blurriness" : blurriness
    }
```

```python
filenames = []

for filename in os.listdir(base_dir):
    if filename.lower().endswith((".jpg", ".png", ".jpeg")):
        info = parse_ccpd_name(filename)
        # if abs(info["angle_h"]) > 15:
        #     continue
        # if abs(info["angle_v"]) > 15:
        #     continue
        # if info["w"] < 80 or info["h"] < 25:
        #     continue
        if int(info["blurriness"]) > 30:
            continue
        # if int(info["brightness"]) < 80:
        #     continue
        # if int(info["brightness"]) > 90:
        #     continue
        if int(info["brightness"]) < 40:
            continue
        if int(info["brightness"]) > 60:
            continue
        filenames.append(filename)
 
        src_path = os.path.join(base_dir, filename)
        dst_path = os.path.join(filter_dir, filename)
        shutil.copy(os.path.join(src_path), os.path.join(dst_path))

print(len(filenames))
```

```python
# paths
source_dir = filter_dir
output_dir = "CCPD2019/splitted/directory/location"

train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

# Create Output Folders
for split in ["train", "valid", "test"]:
    os.makedirs(os.path.join(output_dir, split), exist_ok=True)

# Get All Images
images = os.listdir(source_dir)
random.shuffle(images)

total = len(images)
train_end = int(total * train_ratio)
val_end = train_end + int(total * val_ratio)

splits = {
    "train": images[:train_end],
    "valid": images[train_end:val_end],
    "test": images[val_end:]
}

# Insert Images into Folders
for split, files in splits.items():
    for f in files:
        shutil.move(
            os.path.join(source_dir, f),
            os.path.join(output_dir, split, f)
        )

print("Done!")
```
